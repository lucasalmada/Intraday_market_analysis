```python
import plotly.graph_objects as go
import mplfinance as mpf
import pandas as pd
from datetime import datetime
import talib
```

```python
df = pd.read_csv('C:\\Users\\lucas\\Documents\\Intraday_dados\\WDO\\WDO$_H1_201212030900_202009011700.csv', sep = '\t')
df.columns = ['Date_t','Time','Open','High','Low','Close','Tickvol','Vol','Spread']

df['Date'] = df['Date_t'] +' '+ df['Time']
df['Date'] = df['Date'].apply(lambda x: datetime.strptime(x, '%Y.%m.%d %H:%M:%S'))

df = df.drop(columns = ['Date_t', 'Time'])

df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Tickvol', 'Vol', 'Spread']]
```

```python
df = df[df['Date'] > '2020-08-15']

df = df.set_index('Date')
```

```python
iday = df.loc['2020-08-15 9:00':,:]
mpf.plot(iday,type='candle')
```

```python
candle_names = talib.get_function_groups()['Pattern Recognition']
```

```python
candle_names
```

```python
# extract OHLC 
op = df['Open']
hi = df['High']
lo = df['Low']
cl = df['Close']
# create columns for each pattern
for candle in candle_names:
    # below is same as;
    # df["CDL3LINESTRIKE"] = talib.CDL3LINESTRIKE(op, hi, lo, cl)
    df[candle] = getattr(talib, candle)(op, hi, lo, cl)
```
